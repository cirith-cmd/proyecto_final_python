## Creacion de Variables

 En esta etapa se generan nuevas columnas en base a las columnas iniciales, con el motivo de responder a futuras preguntas (escribir un poco mas con chatgpt)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'..\data\DataScience_salaries_2024.csv')
```

```python
# Salario promedio por nivel de experiencia
# Funcion: Promedio
# Seleccion: Salario
# Agrupacion: Nivel de experiencia
# Correlacion: Duplicar datos segun nivel de experiencia (metodo transform), se uso tambien .round para redondear los valores en usd y que fuera mas claro
df['salario_promedio_segun_experiencia_en_usd'] = df.groupby('experience_level')['salary_in_usd'].transform('mean').round(2)
df.head()
```

```python
# Salario promedio por t√≠tulo de trabajo
# Funcion: Promedio
# Seleccion: Salario
# Agrupacion: Nombre del trabajo
df['salario_promedio_por_titulo_de_trabajo'] = df.groupby('job_title')['salary_in_usd'].transform('mean')
df.head()
```

```python
# Salario promedio por a√±o 
# Funcion: Promedio
# Seleccion: Salario
# Agrupacion: A√±o de trabajo y Tama√±o de la empresa
df['salario_promedio_por_a√±o_y_por_tama√±o_de_la_empresa'] = df.groupby(['work_year', 'company_size'])['salary_in_usd'].transform('mean')
df.head()
```

```python
#diferencia porcentual entre el salario del empleado y el salario promedio de todos los empleados
salario_promedio = df['salary_in_usd'].mean()
df['salario_vs_promedio'] = (df['salary_in_usd'] - salario_promedio) / salario_promedio * 100
df['salario_vs_promedio'] = df['salario_vs_promedio'].round(2)
df.head()
```

```python
#Salario ajustado por nivel de experiencia, para resaltar c√≥mo la experiencia impacta en la remuneraci√≥n.

#primero se crea una columna con la jerarquia o nivel de importancia de los niveles de experiencia
mapa_jerarquico = {
    'EN': 1,
    'MI': 2,
    'SE': 3,
    'EX': 4
}
#Se colocan valores segun correspondencia de la llave del diccionario con los valores de la columna experience_level
df['jerarquia_experiencia'] = df['experience_level'].map(mapa_jerarquico)
#Se calcula el salario ajustado por experiencia
df['salario_ajustado_por_experiencia'] = df['salary_in_usd'] / df['jerarquia_experiencia']
df['salario_ajustado_por_experiencia'] = df['salario_ajustado_por_experiencia'].round(2)
df.head()
```

```python
#Salario promedio segun ubicacion de la empresa

#se agrupa por ubicacion y se calcula el promedio de los grupos formados para la columna salary_in_usd
#ademas se genera la columna salario_promedio_por_locacion segun la correlacion con la columna company_location
df['salario_promedio_por_locacion'] = df.groupby('company_location')['salary_in_usd'].transform('mean')
#diferencia salarial
df['diferencia_salarial_por_ubicacion'] = df['salary_in_usd'] - df['salario_promedio_por_locacion']
df['diferencia_salarial_por_ubicacion'] = df['diferencia_salarial_por_ubicacion'].round(2)
df.head()
```
